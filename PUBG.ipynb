```python
AR = ['AKM','M416','SCAR-L','Groza','M16A4','AUG']
SR = ['Kar98k', 'M24','AWM', 'Win94']
DMR = [ 'Mk14', 'VSS', 'SKS','Mini 14']
SMG = ['Vector','Tommy Gun', 'UMP9','Micro UZI']
LMG = ['M249','DP-28']
CB = ['Crossbow']
Out = ['Down and Out']
BZ = ['Bluezone','death.PlayerMale_A_C']
SG = [ 'S686', 'S12K', 'S1897' ]
HG = ['P1911', 'P92','P18C','R45', 'R1895','death.WeapSawnoff_C']
Thrown  = ['Grenade','death.ProjMolotov_DamageField_C', 'death.Buff_FireDOT_C', 'death.ProjMolotov_C']
Melee  = [ 'Machete', 'Crowbar', 'Sickle','Punch','Pan']
Accident = ['Drown','Falling', 'RedZone']
Vehicle = ['Motorbike (SideCar)', 'Motorbike', 'Pickup Truck', 'Boat','Van','Uaz','Dacia','Buggy','Aquarail','Hit by Car','death.PG117_A_01_C']
```

```python
lis1 = []
lis2 = []
Cate = {}

for x in AR:
    lis1.append("AR")
    lis2.append(x)
for x in SR:
    lis1.append("SR")
    lis2.append(x)
for x in DMR:
    lis1.append("DMR")
    lis2.append(x)
for x in SMG:
    lis1.append("SMG")
    lis2.append(x)
for x in LMG:
    lis1.append("LMG")
    lis2.append(x)
for x in CB:
    lis1.append("CB")
    lis2.append(x)
for x in Out:
    lis1.append("Out")
    lis2.append(x)
for x in BZ:
    lis1.append("BZ")
    lis2.append(x)
for x in SG:
    lis1.append("SG")
    lis2.append(x)
for x in HG:
    lis1.append("HG")
    lis2.append(x)
for x in Thrown:
    lis1.append("Thrown")
    lis2.append(x)
for x in Melee:
    lis1.append("Melee")
    lis2.append(x)
for x in Accident:
    lis1.append("Accident")
    lis2.append(x)
for x in Vehicle:
    lis1.append("Vehicle")
    lis2.append(x)
```

```python
Cate['type'] = lis1
Cate['killed_by'] = lis2
Weapon_cate = pd.DataFrame.from_dict(Cate)
```

```python
kill_match_stats = pd.merge(kill_match_stats,Weapon_cate, left_on=['killed_by'], right_on = ['killed_by'], how = 'left')
```

```python
kill_match_stats
```

```python
def Euclidean_Dist(df1, df2):
    return np.linalg.norm(df1.values - df2.values,axis=1)

pixel_dist = Euclidean_Dist(kill_match_stats[['killer_position_x', 'killer_position_y']],kill_match_stats[['victim_position_x', 'victim_position_y']])
m_dist = pixel_dist*0.0002645833 #pixel to m

kill_match_stats['distance'] = m_dist
```

```python
kill_match_stats[['type','distance']]
```

```python
k = np.array(kill_match_stats[['killer_position_x','killer_position_y']])
v = np.array(kill_match_stats[['victim_position_x','victim_position_y']])

Angle = np.arctan2((k-v)[:,0],(k-v)[:,1])*180/np.pi 
kill_match_stats['angle'] = Angle
del kill_match_stats['victim_position_x']
del kill_match_stats['victim_position_y']

kill_match_stats['time'] = kill_match_stats['time'].floordiv(60)
kill_match_stats['distance'] = kill_match_stats['distance'].floordiv(1)
```

```python
def stacked_bar_plot(df,row):
    r = [i for i  in df.T]
    totals = list(df.sum(axis = 1))
    mod = sys.modules[__name__]
    colors = ['#FF0000','#01DF01','#0101DF','#2EFEC8','#FE2E9A','#2E2E2E','#4B610B','#00FFFF','#A4A4A4','#8A0868','#A9F5BC','#FE9A2E','#61210B','#8258FA']

    df.plot.bar(stacked=True, ax=axes[row,0], edgecolor='black', width=1, color= colors)

    bar_lis = []
    for x in list(i for i in df):
        setattr(mod, 'bar_{}'.format(str(x)), [(a / b) * 100 for a,b in zip(df[str(x)], totals)])
        bar_lis.append('bar_{}'.format(str(x)))

    b_c = {}
    for x,y in zip(bar_lis,colors):
        b_c[x] = y
    
    barWidth = 1
    rows = tuple(i for i  in df.T)

    bot = np.zeros(shape=(len(df)))

    for x in bar_lis:
        a = getattr(mod, x)
        axes[row,1].bar(r, a, bottom = list(bot), color=b_c[x], edgecolor='black',width=barWidth, label= x)
        bot += np.array(a)

    axes[row,1].set_xlabel("Minute")
    axes[row,1].set_ylabel("Mass percentage")
    axes[row,1].legend(loc='upper left', bbox_to_anchor=(1,1), ncol=2)
    return 
```

```python
plt.rcParams["figure.figsize"] =[26,14]

fig, axes = plt.subplots(nrows=2, ncols=2)

pivot_df = kill_match_stats[['type','time']].pivot_table(index='time', columns='type', aggfunc=len, fill_value=0)
stacked_bar_plot(pivot_df,0)

pivot_df =kill_match_stats[kill_match_stats['killed_by'] != 'Down and Out'][['type','time']].pivot_table(index='time', columns='type', aggfunc=len, fill_value=0)
stacked_bar_plot(pivot_df,1)

plt.show()
```
